# 🔹 Q12 (Full Name + Upper/Lower)  User se first name aur last name lo aur:  1)  full name print karo  2) upper case 3)lower case

```python
fname = input("Enter your First name : ")
lname = input("Enter your Last name : ")
name = f"{fname} {lname}"
print(f"Full Name: {name}")
print(f"Upper Case: {name.upper()}")
print(f"Lower Case: {name.lower()}")
```

# 🔹 Q13 (Sentence Formatting) User se aik sentence lo aur usay: 1) title case 2)capitalize  3)length print karo

```python
user = input("Enter a sentence:  ")
print(f"Title Case of Sentence is : {user.title()}")
print(f"Capitalized of Sentence is :{user.capitalize()}")
print(f"Length of Sentence is : {len(user)}")
```

# 🔹 Q14 (Marks + Total) User se 3 subjects ke marks lo aur total print karo using f-string.

```python
m1 = int(input("Enter marks for subject 1: "))
m2 = int(input("Enter marks for subject 2: "))
m3 = int(input("Enter marks for subject 3: "))
total = m1 + m2 + m3
print(f" subject 1 = {m1}\n subject 2 = {m2}\n subject 3 = {m3}")
print(f"Total marks of 3 subjects = {total}")
```
