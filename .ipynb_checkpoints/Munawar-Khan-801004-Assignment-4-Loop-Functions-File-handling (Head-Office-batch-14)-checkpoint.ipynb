**Q1: `for loop` ka use karke 1 se lekar 50 tak saare numbers print karein.**

```python
for number in range(1,51):
 print(number)
```

**Q2: `while loop` ka istemal karte hue 10 se 1 tak reverse counting print karein.**

```python
number = 10
while number >= 1:
    print(number)
    number -= 1
```

**Q3: User se ek number input lein aur us number ka table (multiplication table) print karein.**

```python
number = int(input("Koi number enter karein: "))
for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
```

**Q4: 1 se 100 tak ke beech ke saare Even numbers print karein.**

```python
for num in range (2,101,2):
    print(num)
```

**Q5: Ek list `fruits = ['apple', 'banana', 'cherry']` ke items print karein.**

```python
fruits = ['apple', 'banana', 'cherry']
print("list items are")
for fruits in fruits:
    print(fruits)
```

**Q6: 1 se lekar 20 tak ke saare numbers ka Sum calculate karein.**

```python
total = 0

for number in range(1, 21):
    total += number
print(f"1 se 20 tak ke numbers ka sum jo aya he wo he = {total}")
```

**Q7: User se tab tak input lein jab tak wo '0' enter na kare (While loop).**

```python
number = int(input("enter a number (stop ke liye 0 likhe): "))
while number != 0:
    print(f"Aap ne {number} enter kiya ")
    number = int(input("enter a number (stop ke liye 0 likhe):"))

print("Program stop ho gaya kyu ke ap ne 0 likha he .")
```

**Q8: List `[12, 45, 7, 90]` mein se Maximum number find karein.**

```python
number = [12, 45, 7, 90]
maximum = max(number)

print(f"Maximum number is: {maximum} below is the list")
for number in number:
    print(number)
```

**Q9: String 'Python' ko loop se reverse print karein.**

```python
text = "Python"

for i in range(len(text) - 1, -1, -1):
    print(text[i], end="")
```
