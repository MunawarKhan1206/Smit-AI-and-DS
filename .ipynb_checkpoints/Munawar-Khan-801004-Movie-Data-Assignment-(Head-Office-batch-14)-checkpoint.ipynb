### **Question 36**
(Visualization) 'Drama' aur 'Action' genres ki movies ke 'Gross' revenue ko compare karne ke liye ek Grouped Box Plot ya Violin Plot banayein.

```python
def get_genre(x):
    if isinstance(x, str):
        if 'Action' in x and 'Drama' not in x:
            return 'Action'
        elif 'Drama' in x and 'Action' not in x:
            return 'Drama'
    return None

movies['group'] = movies['Genre'].apply(get_genre)

d = movies.dropna(subset=['group', 'Gross']).copy()

d['Gross'] = pd.to_numeric(d['Gross'], errors='coerce')

plt.figure(figsize=(10,5))
sns.boxplot(x='group', y='Gross', data=d)
plt.yscale('log')
plt.title('Action vs Drama Gross')
plt.show()
```

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
movies['imdb_scaled'] = movies['IMDB_Rating'] * 10

top_movies = movies.nlargest(5, 'IMDB_Rating')

names = top_movies['Series_Title'].values
meta_score = top_movies['Meta_score'].values
imdb_score = top_movies['imdb_scaled'].values

x = np.arange(len(names))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(x - width/2, meta_score, width, label='Meta Score')
ax.bar(x + width/2, imdb_score, width, label='IMDB (scaled)')

ax.set_xlabel('Movies')
ax.set_ylabel('Score')
ax.set_title('Top 5 Movies: Meta vs IMDB')
ax.set_xticks(x)
ax.set_xticklabels(names, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
