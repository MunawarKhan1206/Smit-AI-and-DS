# Reusable Functions for Data Cleaning and EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

def dataclean(cd):
    cd.rename(columns=lambda x: x.strip(), inplace=True)
    
    obj_cols = cd.select_dtypes(include=['object']).columns
    for column in obj_cols:
        cd[column] = cd[column].astype(str).str.strip().replace({'?': np.nan, 'nan': np.nan, 'None': np.nan})
        
    for column in cd.columns:
        if cd[column].hasnans:
            if np.issubdtype(cd[column].dtype, np.number):
                cd[column].fillna(cd[column].median(), inplace=True)
            else:
                cd[column].fillna(cd[column].mode()[0], inplace=True)
                
    return cd.drop_duplicates().reset_index(drop=True)

def show_info(cd):
    print(f"Shape: {cd.shape}\n\nInfo:")
    cd.info()
    print(f"\nMissing Values:\n{cd.isna().sum()}\n\nDuplicate Rows: {cd.duplicated().sum()}")

def plot_numeric(cd, col_name):
    fig, (ax_hist, ax_box) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
    
    sns.histplot(data=cd, x=col_name, kde=True, ax=ax_hist)
    ax_hist.set_title(f"{col_name} HIstogram")
    
    sns.boxplot(data=cd, x=col_name, ax=ax_box)
    ax_box.set_title(f"{col_name} Boxplot")
    
    plt.tight_layout()
    plt.show()

def plot_categorical(cd, col_name):
    plt.figure(figsize=(10, 4))
    
    val_counts = cd[col_name].value_counts()
    order = val_counts.index[:30] if len(val_counts) > 30 else val_counts.index
    
    sns.countplot(data=cd, x=col_name, order=order)
    plt.title(f"Countplot of {col_name}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def check_outliers_iqr(cd):
    results = []
    num_cols = cd.select_dtypes(include=['int64', 'float64']).columns
    
    for column in num_cols:
        q1, q3 = cd[column].quantile([0.25, 0.75])
        iqr_val = q3 - q1
        lower_bound = q1 - 1.5 * iqr_val
        upper_bound = q3 + 1.5 * iqr_val
        
        num_outliers = cd[(cd[column] < lower_bound) | (cd[column] > upper_bound)].shape[0]
        
        results.append({
            'Column': column,
            'Q1': q1,
            'Q3': q3,
            'IQR': iqr_val,
            'Outliers Count': num_outliers
        })
        
    return pd.DataFrame(results)

def plot_heatmap(cd):
    num_cd = cd.select_dtypes(include=['int64', 'float64'])
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(num_cd.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Heatmap")
    plt.tight_layout()
    plt.show()
```

```python
loan = pd.read_csv('loan-train.csv')
```

```python
show_info(loan)
```

```python
loan.head()
```

```python
loan = dataclean(loan)
loan.describe()
```

```python
check_outliers_iqr(loan)
```

```python
plot_heatmap(loan)
```

```python
for column in loan.select_dtypes(include=['int64', 'float64']).columns:
    plot_numeric(loan, column)
```

```python
for column in loan.select_dtypes(include=['object']).columns:
    plot_categorical(loan, column)
```

```python
loan.to_csv('folder/loan.csv', index=False)
```
