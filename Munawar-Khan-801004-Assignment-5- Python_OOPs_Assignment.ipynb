**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

```python
class Student:
    def __init__(self,name,roll_number):
        self. __name = name
        self. __roll_number = roll_number
    def get_name(self):
        return self. __name
    def set_name(self,new_name):
        self. __name = new_name
    def get_roll_number(self):
        return self. __roll_number
    def set_roll_number(self,new_roll_number):
        self. __roll_number = new_roll_number
    def show(self):
        print(f"Your name is {self.__name} and roll number is {self. __roll_number}")
obj = Student("Munawar Khan","801004")
obj.show()
obj.get_name()
obj.set_name("Saim Raza")
obj.get_name()
obj.set_roll_number(401)
obj.get_roll_number()
```

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class Account:
    def __init__(self,balance):
        self.__balance = balance
    def deposit(self,amount):
        if amount <= 0:
            print("Invalid Amount Entered")
        else:
            self.__balance +=amount
            print(f"Amount Sucessfully deposited. Your new balance is {self.__balance + amount}")
    def withdraw(self,amount):
        if amount > self.__balance:
            print("Insufficent amount is entered")
        else:
            self.__balance -=amount
            print(f"Amount withdrawed. Your new balance is {self.__balance - amount}")
obj = Account(5000)
obj.deposit(1000)
obj.withdraw(900)
obj.withdraw(1100)
        
```

**Q4:** Protected variable (`_`) aur Private variable (`__`) ke beech kya farq hota hai? Aik simple code ke zariye dono ka use-case dikhayein.

```python
class Parent:
    def __init__(self):
        self._protected_val = "Protected Data"  
        self.__private_val = "Private Data"     
class Child(Parent):
    def show_data(self):
        print(f"Child node accessing protected: {self._protected_val}") 
        try:
            print(self.__private_val)
        except AttributeError:
            print("Child node cannot access private variable")
obj = Child()
obj.show_data()

print("\n--- external node ---")
print(f"external node accessing protected: {obj._protected_val}")
try:
    print(obj.__private_val)
except AttributeError:
    print("Child node cannot access private variable")
print(f"Accessing private node via Name Mangling: {obj._Parent__private_val}")
```

**Q6:** Single Inheritance ka use karte hue aik parent class `Vehicle` (with brand attribute) aur child class `Car` (with model attribute) banayein. Car ka object bana kar dono attributes print karein.

```python
class Vehicle:
    def __init__(self,brand):
        self.brand = brand
    def show(self):
        print(f"Your vehicle brand is {self.brand}")
class Car(Vehicle):
    def __init__(self, brand,model):
        super().__init__(brand)
        self.model = model
    def show(self):
        super().show()
        print(f"Your vehicle model is {self.model}")
obj = Car("BMW","M4")
obj.show()
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
class Shape:
    def __init__(self,color):
        self.color = color
    def intro(self):
        print(f"The color of the shape is {self.color}")
class Circle(Shape):
    def __init__(self, color,radius):
        super().__init__(color)
        self.radius = radius
    def area(self):
        print(f"The area of circle is {3.14 * self.radius * self.radius}")
class Square(Shape):
    def __init__(self, color,sides):
        super().__init__(color)
        self.sides = sides
    def area(self):
        print(f"The area of square is {self.sides * self.sides}")
obj = Circle("yellow",23)
obj.intro()
obj.area()
obj2 = Square("green",14)
obj2.intro()
obj2.area()
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Number:
    def __init__(self,n) -> None:
        self.n = n
    def __add__(self, other):
        return self.n + other.n
n = Number(11)
m = Number(9)
print(n + m)
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

```python
from abc import ABC,abstractmethod
class Computer(ABC):
    @abstractmethod
    def process(self):
        pass
class Laptop(Computer):
    def __init__(self):
        print("I Am a laptop")
    def process(self):
        print("I am a portable thing")
class Desktop(Computer):
    def __init__(self):
        print("I am a Desktop")
    def process(self):
        print("I Am not portable")
obj = Laptop()
obj.process()
obj2 = Desktop()
obj2.process()
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
class Car(ABC):
    def start_engine(self):
        print("Car is starting")
    @abstractmethod
    def drive(self):
        pass
class ElectricCar(Car):
    def drive(self):
        print("The car is being driven properly")
obj = ElectricCar()
obj.start_engine()
obj.drive()
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC,abstractmethod

class Payment(ABC):
    def __init__(self,amount):
        self.amount = amount
    @abstractmethod
    def processPayment(self):
        '''Every payment method must define its own'''
        pass
    def paymentRecipt(self):
        print(f"Payment recieved successfully {self.amount}")
class CreditCardPayment(Payment):
    def __init__(self, amount,Number,Cardname):
        super().__init__(amount)
        self.Number = Number
        self.Cardname = Cardname
    def processPayment(self):
        print(f"Processing your transaction for card number {self.Number}")
        print(f"Card Holder name is {self.Cardname}")
    def paymentRecipt(self):
        super().paymentRecipt()
        print(f"Payment is processed by card. The holder name is {self.Cardname}")
class UpiPayment(Payment):
    def __init__(self, amount,digital_id):
        super().__init__(amount)
        self.digital_id = digital_id
    def processPayment(self):
        print(f"Initiating UPI payment request via ID: {self.digital_id}...")
        print("Waiting for secure MPIN authentication...")
        print(f"Rs. {self.amount} successfully transferred via Digital gateway.")
    def paymentRecipt(self):
        super().paymentRecipt()
        print(f"This payment is processed by  id-> {self.digital_id}")
        
print("--- Transaction 1: Shopping Mall ---")
cardp = CreditCardPayment(14500, "45678911121314", "Munawar Khan")
cardp.processPayment()
cardp.paymentRecipt()

print("--- Transaction 2: Online QR Scanner ---")
digitalp = UpiPayment(1200, "munawar@paypal")
digitalp.processPayment()
digitalp.paymentRecipt()
```
