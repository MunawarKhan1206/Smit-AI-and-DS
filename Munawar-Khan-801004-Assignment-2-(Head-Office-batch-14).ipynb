# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
mark = int(input("Enter marks: "))

if mark >= 90:
    print("Grade A")
elif mark >= 80:
    print("Grade B")
elif mark >= 70:
    print("Grade C")
elif mark >= 60:
    print("Grade D") 
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
numday = int(input("Enter (1-7) Number for Days: "))
if numday == 1:
    print("Monday")
elif numday == 2:
    print("Tuesday")
elif numday == 3:
    print("Wednesday")
elif numday == 4:
    print("Thursday")
elif numday == 5:
    print("Friday")
elif numday == 6:
    print("Saturday")
elif numday ==7:
    print("Sunday")
else:
    print("Enter 1-7 number only")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
monnum = int(input("Enter Number (1-12) of Month"))
if monnum == 12 or monnum == 1 or monnum == 2:
    print("Winter")
elif monnum == 3 or monnum == 4 or monnum == 5:
    print("Spring")
elif monnum == 6 or monnum == 7 or monnum == 8:
    print("Summer")
else:
    print("Autumn")
```

# User se 2 numbers lo:

# pehla > dusra → "First is greater"
# pehla < dusra → "Second is greater"
# warna → "Both equal"

```python
num1 = float(input("Enter num 1 : "))
num2 = float(input("Enter num 2 : "))

if num1 > num2:
    print(f"First is greater: {num1} > {num2}")
elif num2 > num1:
    print(f"Second is greater: {num2} > {num1}")
else:
    print(f"Both equal: {num1} = {num2}")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
operator = input("Enter a Operator")
if operator == "+":
    print("Addition")
elif operator == "-":
    print("Subratactiom")
elif operator == "/":
    print("Divison")
elif operator == "*":
    print("Multiplication")
else:
    print("INVALID OPERATOR")
```
